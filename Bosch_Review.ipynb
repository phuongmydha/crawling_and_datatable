```python
import pandas as pd
import regex as re
from underthesea import word_tokenize, pos_tag
from pyvi import ViTokenizer
import json
from py_vncorenlp import VnCoreNLP
from transformers import AutoTokenizer
import unicodedata

class TextPreprocessor:
    def __init__(self):
        # Khởi tạo py_vncorenlp
        # Tải mô hình nếu chưa có
        self.rdrsegmenter = py_vncorenlp.VnCoreNLP(save_dir='/content/vncorenlp', annotators=["wseg", "pos", "ner", "parse"])

        # Khởi tạo PhoBERT tokenizer
        self.phobert_tokenizer = AutoTokenizer.from_pretrained("vinai/phobert-base")

        # Load từ điển teen code và từ viết tắt
        with open('/content/drive/MyDrive/BoschReview/teencode.js', 'r', encoding='utf-8') as f:
            self.teen_code_dict = json.load(f)

        # Load danh sách stopwords
        with open('/content/drive/MyDrive/BoschReview/vietnamese-stopwords.txt', 'r', encoding='utf-8') as f:
            self.stopwords = set(f.read().splitlines())

    def normalize_unicode(self, text):
        """Chuẩn hóa unicode, xử lý các dấu tiếng Việt"""
        text = unicodedata.normalize('NFKC', text)
        return text

    def replace_teen_code(self, text):
        """Thay thế teen code và từ viết tắt"""
        words = text.split()
        normalized_words = [self.teen_code_dict.get(word.lower(), word) for word in words]
        return ' '.join(normalized_words)

    def remove_special_characters(self, text):
        """Xóa ký tự đặc biệt nhưng giữ lại dấu tiếng Việt"""
        text = re.sub(r'[^\s\wáàảãạăắằẳẵặâấầẩẫậéèẻẽẹêếềểễệóòỏõọôốồổỗộơớờởỡợíìỉĩịúùủũụưứừửữựýỳỷỹỵđ_]',' ',text)
        return text

    def standardize_marks(self, text):
        """Chuẩn hóa dấu câu và khoảng trắng"""
        text = re.sub(r'\s+', ' ', text)
        text = text.strip()
        return text

    def remove_emoji(self, text):
        """Xóa emoji và các ký tự đặc biệt"""
        emoji_pattern = re.compile("["
            u"\U0001F600-\U0001F64F"  # emoticons
            u"\U0001F300-\U0001F5FF"  # symbols & pictographs
            u"\U0001F680-\U0001F6FF"  # transport & map symbols
            u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
            u"\U00002702-\U000027B0"
            u"\U000024C2-\U0001F251"
            "]+", flags=re.UNICODE)
        return emoji_pattern.sub(r'', text)

    def word_tokenize_wrapper(self, text, method='vncorenlp'):
        """Tách từ với nhiều phương pháp khác nhau"""
        if method == 'vncorenlp':
            try:
                # Sử dụng word segmentation từ py_vncorenlp
                annotated_text = self.rdrsegmenter.word_segment(text)
                # annotated_text trả về list các từ đã được tách
                return ' '.join(annotated_text)
            except Exception as e:
                print(f"Error in word tokenization: {e}")
                return text
        elif method == 'underthesea':
            return ' '.join(word_tokenize(text))
        elif method == 'phobert':
            tokens = self.phobert_tokenizer.tokenize(text)
            return ' '.join(tokens)
        else:
            return ViTokenizer.tokenize(text)

    def remove_stopwords(self, text):
        """Loại bỏ stopwords"""
        words = text.split()
        filtered_words = [word for word in words if word.lower() not in self.stopwords]
        return ' '.join(filtered_words)

    def remove_urls(self, text):
        """Loại bỏ các URL khỏi văn bản"""
        # Pattern để nhận diện URL
        url_pattern = re.compile(
            r'(https?:\/\/)?'  # protocol (http/https)
            r'([\da-z\.-]+)'   # domain
            r'\.([a-z\.]{2,6})'  # top-level domain
            r'([\/\w \.-]*)*'   # path, parameters
            r'\/?\S*'           # trailing characters
        )

        # Pattern để nhận diện các link facebook cụ thể
        fb_pattern = re.compile(
            r'(?:(?:http|https):\/\/)?'
            r'(?:www.)?'
            r'facebook.com\/'
            r'[a-zA-Z0-9\.]*'
        )

        # Pattern cho các dạng rút gọn URL phổ biến
        short_url_pattern = re.compile(
            r'(?:(?:http|https):\/\/)?'
            r'(?:bit\.ly|t\.co|goo\.gl|tinyurl\.com)'
            r'\/\S*'
        )

        # Loại bỏ các URL
        text = url_pattern.sub('', text)
        text = fb_pattern.sub('', text)
        text = short_url_pattern.sub('', text)

        # Loại bỏ các từ www. còn sót lại
        text = re.sub(r'www\S+', '', text)

        # Xử lý khoảng trắng dư thừa
        text = re.sub(r'\s+', ' ', text)
        return text.strip()

    def preprocess_text(self, text, tokenize_method='vncorenlp'):
        """Pipeline xử lý văn bản hoàn chỉnh"""
        if not isinstance(text, str):
            return ''

        # Các bước xử lý
        text = self.remove_urls(text)
        text = text.lower()
        text = self.normalize_unicode(text)
        text = self.remove_emoji(text)
        text = self.remove_special_characters(text)
        text = self.replace_teen_code(text)
        text = self.standardize_marks(text)
        text = self.word_tokenize_wrapper(text, method=tokenize_method)
        text = self.remove_stopwords(text)

        return text

def process_bosch_reviews(df):
    # Khởi tạo text processor
    processor = TextPreprocessor()

    # Tạo bản sao của DataFrame
    df_processed = df.copy()

    # Xử lý các comment
    df_processed['comment_cleaned'] = df_processed['comment'].apply(
        lambda x: processor.preprocess_text(x, tokenize_method='vncorenlp')
    )

    df_processed['post_content_cleaned'] = df_processed['post_content'].apply(
        lambda x: processor.preprocess_text(x, tokenize_method='vncorenlp')
    )

    return df_processed

# Sử dụng
df_bosch_processed = process_bosch_reviews(df_bosch)

# Lưu kết quả
df_bosch_processed
```

```python
### Version full
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer
import numpy as np
from tqdm import tqdm
import logging
from typing import Dict, List, Optional, Union
from dataclasses import dataclass
import pandas as pd

# Cấu hình logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

@dataclass
class SentimentResult:
    label: str
    confidence: float
    scores: Dict[str, float]

class SentimentAnalyzer:
    def __init__(self, config: Optional[Dict] = None):
        # Default hyperparameters
        self.config = {
            # Model parameters
            'num_labels': 3,
            'hidden_dropout_prob': 0.1,
            'attention_probs_dropout_prob': 0.1,

            # Tokenization parameters
            'max_length': 256,
            'padding': True,
            'truncation': True,

            # Aspect analysis parameters
            'context_window_size': 100,
            'confidence_threshold': 0.4,
            'min_comment_length': 5,

            # Training parameters
            'learning_rate': 2e-5,
            'batch_size': 16,
            'num_epochs': 3,
            'warmup_steps': 500,
            'max_grad_norm': 1.0,
            'weight_decay': 0.01,

            # Aspect extraction parameters
            'min_aspect_mentions': 1,
            'aspect_context_overlap': 0.3,

            # Aspect categories và keywords
            'aspects': {
            'price': [
                'giá', 'đắt', 'rẻ', 'tiền', 'chi phí', 'khuyến mãi', 'giảm giá',
                'đồng', 'vnđ', 'k', 'triệu', 'giá cả', 'giá thành', 'giá tiền'
            ],
            'quality': [
                'chất lượng', 'tốt', 'xấu', 'kém', 'được', 'bền', 'hỏng',
                'ổn định', 'chắc chắn', 'độ bền', 'hiệu quả', 'hiệu năng',
                'hoạt động', 'vận hành'
            ],
            'service': [
                'dịch vụ', 'bảo hành', 'nhân viên', 'tư vấn', 'hỗ trợ',
                'chăm sóc', 'phục vụ', 'bảo trì', 'sửa chữa', 'thái độ'
            ],
            'brand': [
                'bosch', 'thương hiệu', 'hãng', 'uy tín', 'tin cậy',
                'danh tiếng', 'nổi tiếng', 'đức', 'germany'
            ] }
        }

        # Update với custom config
        if config:
            self.config.update(config)

        # Khởi tạo device
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        logger.info(f"Using device: {self.device}")

        try:
            # Khởi tạo tokenizer và model
            self.tokenizer = AutoTokenizer.from_pretrained(
                "vinai/phobert-base",
                use_auth_token="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"
            )

            self.model = AutoModelForSequenceClassification.from_pretrained(
                "vinai/phobert-base",
                num_labels=self.config['num_labels'],
                hidden_dropout_prob=self.config['hidden_dropout_prob'],
                attention_probs_dropout_prob=self.config['attention_probs_dropout_prob']
            ).to(self.device)

            self.model.eval()

        except Exception as e:
            logger.error(f"Error initializing model: {str(e)}")
            raise

    def predict_sentiment(self, text: str) -> Optional[SentimentResult]:
        """
        Dự đoán sentiment cho một đoạn văn bản.

        Args:
            text: Đoạn văn bản cần phân tích

        Returns:
            SentimentResult object hoặc None nếu có lỗi
        """
        try:
            # Validate input
            if not isinstance(text, str) or len(text) < self.config['min_comment_length']:
                return None

            # Tokenization
            inputs = self.tokenizer(
                text,
                padding=self.config['padding'],
                truncation=self.config['truncation'],
                max_length=self.config['max_length'],
                return_tensors="pt"
            ).to(self.device)

            # Prediction
            with torch.no_grad():
                outputs = self.model(**inputs)
                predictions = torch.softmax(outputs.logits, dim=-1)
                scores = predictions.cpu().numpy()[0]

            # Check confidence threshold
            confidence = np.max(scores)
            if confidence < self.config['confidence_threshold']:
                return None

            # Get label
            label = ['Negative', 'Neutral', 'Positive'][np.argmax(scores)]

            return SentimentResult(
                label=label,
                confidence=float(confidence),
                scores={
                    'negative': float(scores[0]),
                    'neutral': float(scores[1]),
                    'positive': float(scores[2])
                }
            )

        except Exception as e:
            logger.error(f"Error in sentiment prediction: {str(e)}")
            return None

    def extract_aspects(self, text: str) -> Dict[str, List[str]]:
        """
        Trích xuất các aspect từ văn bản.

        Args:
            text: Văn bản cần phân tích

        Returns:
            Dict với key là tên aspect và value là list các mention
        """
        found_aspects = {}
        text = text.lower()

        for aspect, keywords in self.config['aspects'].items():
            mentions = []
            for keyword in keywords:
                if keyword in text:
                    mentions.append(keyword)
            if len(mentions) >= self.config['min_aspect_mentions']:
                found_aspects[aspect] = mentions

        return found_aspects

    def _get_aspect_context(
        self,
        text: str,
        mentions: List[str],
        window_size: int,
        overlap_threshold: float
    ) -> str:
        """
        Lấy context window cho các mention của một aspect.

        Args:
            text: Văn bản gốc
            mentions: List các mention cần lấy context
            window_size: Kích thước cửa sổ context
            overlap_threshold: Ngưỡng overlap cho phép

        Returns:
            Đoạn văn bản context đã được ghép
        """
        contexts = []
        for mention in mentions:
            start_idx = text.lower().find(mention)
            if start_idx != -1:
                start = max(0, start_idx - window_size)
                end = min(len(text), start_idx + len(mention) + window_size)
                context = text[start:end]
                contexts.append((start, end, context))

        # Merge overlapping contexts
        merged_contexts = []
        contexts.sort(key=lambda x: x[0])

        current_context = contexts[0]
        for next_context in contexts[1:]:
            overlap = (current_context[1] - next_context[0]) / \
                     (current_context[1] - current_context[0])

            if overlap > overlap_threshold:
                current_context = (
                    current_context[0],
                    max(current_context[1], next_context[1]),
                    text[current_context[0]:max(current_context[1], next_context[1])]
                )
            else:
                merged_contexts.append(current_context)
                current_context = next_context

        merged_contexts.append(current_context)

        return " ".join(context[2] for context in merged_contexts)

    def analyze_aspect_sentiment(self, text: str) -> Dict[str, SentimentResult]:
        """
        Phân tích sentiment cho từng aspect trong văn bản .

        Args:
            text: Văn bản cần phân tích

        Returns:
            Dict với key là tên aspect và value là SentimentResult
        """
        aspects = self.extract_aspects(text)
        results = {}

        for aspect, mentions in aspects.items():
            context = self._get_aspect_context(
                text,
                mentions,
                window_size=self.config['context_window_size'],
                overlap_threshold=self.config['aspect_context_overlap']
            )
            results[aspect] = self.predict_sentiment(context)

        return results
```

```python
def analyze_bosch_comments(df: pd.DataFrame) -> pd.DataFrame:
    """
    Phân tích sentiment cho dataset với comment và post content đã được làm sạch.

    Args:
        df: DataFrame chứa comment_cleaned và post_content_cleaned

    Returns:
        DataFrame chứa kết quả phân tích
    """
    analyzer = SentimentAnalyzer()
    results = []

    for idx, row in tqdm(df.iterrows(), total=len(df)):
        comment = row['comment_cleaned']
        post_content = row['post_content_cleaned']

        # Skip nếu cả comment và post_content đều trống
        if (not isinstance(comment, str) or not comment.strip()) and \
           (not isinstance(post_content, str) or not post_content.strip()):
            continue

        # Phân tích comment
        comment_sentiment = None
        comment_aspects = None
        if isinstance(comment, str) and comment.strip():
            comment_sentiment = analyzer.predict_sentiment(comment)
            comment_aspects = analyzer.analyze_aspect_sentiment(comment)

        # Phân tích post content
        post_sentiment = None
        post_aspects = None
        if isinstance(post_content, str) and post_content.strip():
            post_sentiment = analyzer.predict_sentiment(post_content)
            post_aspects = analyzer.analyze_aspect_sentiment(post_content)

        # Kết hợp context để phân tích
        combined_text = ""
        if isinstance(comment, str) and comment.strip():
            combined_text += comment + " "
        if isinstance(post_content, str) and post_content.strip():
            combined_text += post_content

        combined_sentiment = None
        combined_aspects = None
        if combined_text.strip():
            combined_sentiment = analyzer.predict_sentiment(combined_text)
            combined_aspects = analyzer.analyze_aspect_sentiment(combined_text)

        main_aspect = get_main_aspect_simple(combined_aspects)

        # Tạo kết quả phân tích
        analysis = {
            'comment_id': row.get('comment_id'),
            'post_id': row.get('post_id'),

            # Comment analysis
            'original_comment': row.get('comment'),
            'cleaned_comment': comment,
            'comment_sentiment': comment_sentiment,
            'comment_aspects': comment_aspects,

            # Post content analysis
            'original_post_content': row.get('post_content'),
            'cleaned_post_content': post_content,
            'post_sentiment': post_sentiment,
            'post_aspects': post_aspects,

            # Combined analysis
            'combined_sentiment': combined_sentiment,
            'combined_aspects': combined_aspects,
            'main_aspect': main_aspect,

            # Meta analysis
            'sentiment_agreement': _check_sentiment_agreement(
                comment_sentiment,
                post_sentiment
            ),
            'context_relevance': _calculate_context_relevance(
                comment,
                post_content
            ),
            'aspect_overlap': _calculate_aspect_overlap(
                comment_aspects,
                post_aspects
            )

        }

        results.append(analysis)

    return pd.DataFrame(results)

def _check_sentiment_agreement(
    comment_sentiment: Optional[SentimentResult],
    post_sentiment: Optional[SentimentResult]
) -> Optional[float]:
    """Kiểm tra độ tương đồng giữa sentiment của comment và post"""
    if not comment_sentiment or not post_sentiment:
        return None

    if comment_sentiment.label == post_sentiment.label:
        return 1.0

    # Tính cosine similarity giữa các scores
    comment_scores = np.array([
        comment_sentiment.scores['negative'],
        comment_sentiment.scores['neutral'],
        comment_sentiment.scores['positive']
    ])
    post_scores = np.array([
        post_sentiment.scores['negative'],
        post_sentiment.scores['neutral'],
        post_sentiment.scores['positive']
    ])

    similarity = np.dot(comment_scores, post_scores) / \
                (np.linalg.norm(comment_scores) * np.linalg.norm(post_scores))

    return float(similarity)

def get_main_aspect_simple(aspects: Optional[Dict]) -> Optional[str]:
    """
    Xác định aspect chính chỉ dựa trên số lần xuất hiện.
    """
    if not aspects:
        return None

    try:
        # Đếm số lần xuất hiện của mỗi aspect
        aspect_counts = {}
        for aspect in aspects:
            if isinstance(aspect, str):  # Đảm bảo aspect là string
                aspect_counts[aspect] = aspect_counts.get(aspect, 0) + 1

        # Trả về aspect xuất hiện nhiều nhất
        if aspect_counts:
            return max(aspect_counts.items(), key=lambda x: x[1])[0]

    except Exception as e:
        print(f"Error in get_main_aspect_simple: {str(e)}")
        return None

    return None


def _calculate_context_relevance(
    comment: Optional[str],
    post_content: Optional[str]
) -> Optional[float]:
    """Tính độ liên quan giữa comment và post content"""
    if not comment or not post_content:
        return None

    # Có thể sử dụng các phương pháp như:
    # - TF-IDF similarity
    # - Semantic similarity với sentence embeddings
    # - Keyword overlap
    # Ở đây ta dùng simple word overlap
    comment_words = set(comment.lower().split())
    post_words = set(post_content.lower().split())

    if not comment_words or not post_words:
        return 0.0

    overlap = len(comment_words.intersection(post_words))
    union = len(comment_words.union(post_words))

    return overlap / union if union > 0 else 0.0

def _calculate_aspect_overlap(
    comment_aspects: Optional[Dict],
    post_aspects: Optional[Dict]
) -> Optional[float]:
    """Tính độ overlap giữa các aspects được đề cập trong comment và post"""
    if not comment_aspects or not post_aspects:
        return None

    comment_aspect_set = set(comment_aspects.keys())
    post_aspect_set = set(post_aspects.keys())

    if not comment_aspect_set or not post_aspect_set:
        return 0.0

    overlap = len(comment_aspect_set.intersection(post_aspect_set))
    union = len(comment_aspect_set.union(post_aspect_set))

    return overlap / union if union > 0 else 0.0

# Sử dụng
df_sentiment_collected = analyze_bosch_comments(df_bosch_processed)

# In kết quả tổng quan
print("\nSentiment Distribution:")
print(df_sentiment_collected['combined_sentiment'].apply(lambda x: x.label if x else None).value_counts())

print("\nAspect Coverage:")
aspect_coverage = df_sentiment_collected['combined_aspects'].apply(
    lambda x: list(x.keys()) if x else []
).explode().value_counts()
print(aspect_coverage)

print("\nAverage Context Relevance:",
      df_sentiment_collected['context_relevance'].mean())

print("\nAverage Sentiment Agreement:",
      df_sentiment_collected['sentiment_agreement'].mean())
```
